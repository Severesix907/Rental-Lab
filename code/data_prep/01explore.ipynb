## Data Exploration

Before cleaning our dataset, let's explore the features that our dataset contains, as well as visuals.

```python
# import required modules for this project

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading Dataset
df = pd.read_excel("../../data/raw/realestate.xlsx")
#Going to print out the first 5 rows of the Dataset.
df.head(5)
```

```python
#Find out the datatypes of each column
df.info(type)
```

```python
#print out the summary statistics (mean, median, max, min, count) of all numeric columns
df.describe()
```

# Univariate Exploratory Analysis

Let's explore the distribution of single variables in this dataset.

```python
#Plotting "House price of unit area" with Histplot
sns.histplot(df["Y house price of unit area"], kde=True)
plt.title("House Price")
plt.xlabel("Price Unit Area")
plt.ylabel("Houses amount")
plt.show()
```

```python
#Plotting boxplots for "House price of unit area"
sns.boxplot(df["Y house price of unit area"])
plt.xlabel("Price of unit")
plt.show()
```

```python
#Plotting histogram of "house age" using sns here
sns.histplot(df["X2 house age"], kde = True)
plt.xlabel("House age")
plt.ylabel('Count')
```

```python
#Plotting boxplot of "House age" using sns
sns.boxplot(df['X2 house age'])
plt.title("House Age")
```

```python
#Plotting histogram of "Number of convenience stores"
sns.histplot(df['X4 number of convenience stores'], kde = True)
plt.title("The Number of Convenience Store")
plt.xlabel("Convenience Store")
```

```python
#Plotting boxplot of "number of convenience stores" using sns here
sns.boxplot(df['X4 number of convenience stores'])
plt.title("Number of Convenience Store")
```

# Bivariate Exploratory Analysis

Let's explore the relationship of multiple variables in this dataset.

```python
# scatter plot of house age & price of unit area
sns.scatterplot(data=df, x="X2 house age", y="Y house price of unit area")
plt.xlabel("House Age")
plt.ylabel("House price")
plt.show()
```

```python
#Plotting scatter plot of "X4 number of convenience stores" and "Y house price of unit area"
sns.scatterplot(df['X4 number of convenience stores'])
sns.scatterplot(df['Y house price of unit area'])
plt.ylabel("Number of Stores")
plt.xlabel('Price of house per unit')
```

```python
# plot a correlation heat map of selected columns in the dataset
selected_cols = ["X2 house age", "X4 number of convenience stores", "Y house price of unit area"]
mask = np.triu(np.ones_like(df[selected_cols].corr(), dtype=bool))
sns.heatmap(df[selected_cols].corr(), annot=True, mask=mask)
plt.show()
```
